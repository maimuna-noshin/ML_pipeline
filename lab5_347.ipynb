#Filter method
##Pearson correlation Method

```python
correlation_matrix = x.corr()
```

features having corelation more than 87% is considered as redundant feature and one of them is dropped.

```python
# Set a threshold to remove highly correlated features
threshold = 0.80
corr_features = set()
```

```python
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            corr_features.add(colname)
```

```python
df_corr = x.drop(columns=corr_features)
print(f"After Correlation Filter: {df_corr.shape}")
```

#ReliefF (Relief Feature Selection Algorithm)

Assigns feature scores based on how well a feature distinguishes between instances that are close to each other.
Works well for noisy and high-dimensional data.
Perfect for this dataset

```python
pip install skrebate
```

```python
from skrebate import ReliefF
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=10000, random_state=42)
x = pd.DataFrame(x).astype('float32')  # Convert to float32 to save memory

# Sample a subset if still too large
x_sample = x.sample(n=5000, random_state=42)
y_sample = y[x_sample.index]

# Convert x_sample to NumPy array
x_sample_np = x_sample.values


y_sample_np = np.array(y_sample)


relieff = ReliefF(n_neighbors=5)

# Fit ReliefF on reduced data
relieff.fit(x_sample_np, y_sample_np)


feature_scores = relieff.feature_importances_
```

```python

scores_df = pd.DataFrame({
    'Feature': x_sample.columns,
    'Score': feature_scores
})
scores_df = scores_df.sort_values(by="Score", ascending=False)
# Normalize scores
scaler = MinMaxScaler()
scores_df['Normalized Score'] = scaler.fit_transform(scores_df[['Score']])


print(scores_df)
```

#MRMR
mRMR is a feature selection technique that selects features with high relevance to the target variable (maximizing mutual information with the target) while minimizing redundancy among the selected features (minimizing mutual information between features).

```python
!pip install mrmr-selection
from mrmr import mrmr_classif
```

```python
x_sample = x.sample(n=1000, random_state=42)
y_sample = y[x_sample.index]
feature_names = x_sample.columns.tolist()
selected_features = mrmr_classif(X=x_sample, y=y_sample, K=8)
# Simulate scores: 1.0 for selected features, 0.0 for others
feature_scores = [1.0 if feat in selected_features else 0.0 for feat in feature_names]
scores_df = pd.DataFrame({
    'Feature': feature_names,
    'Score': feature_scores
})
scores_df = scores_df.sort_values(by="Score", ascending=False)
scaler = MinMaxScaler()
scores_df['Normalized Score'] = scaler.fit_transform(scores_df[['Score']])


print("Selected Features:", selected_features)
print("\nScores DataFrame:")
print(scores_df)
```

#Wrapper
##SFS
SFS is a greedy algorithm that iteratively adds or removes features to find the optimal subset.
It works by evaluating the performance of a machine learning model as features are added or removed.It uses a scoring metric (e.g., accuracy, precision, recall) to evaluate the model's performance.






```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
```

```python
x_sample = x.sample(n=1000, random_state=42)
y_sample = y[x_sample.index]
estimator = LogisticRegression(max_iter=1000, random_state=42)
sfs = SequentialFeatureSelector(
    estimator=estimator,
    n_features_to_select=10,
    direction='forward',
    scoring='accuracy',
    cv=5,
    n_jobs=-1
)
sfs.fit(x_sample, y_sample)
selected_features_mask = sfs.get_support()  # Boolean mask of selected features
selected_features = x_sample.columns[selected_features_mask].tolist()
feature_names = x_sample.columns.tolist()
feature_scores = [1.0 if feat in selected_features else 0.0 for feat in feature_names]
scores_df = pd.DataFrame({
    'Feature': feature_names,
    'Score': feature_scores
})
scores_df = scores_df.sort_values(by="Score", ascending=False)

scaler = MinMaxScaler()
scores_df['Normalized Score'] = scaler.fit_transform(scores_df[['Score']])

print("Selected Features:", selected_features)
print("\nScores DataFrame:")
print(scores_df)
```

##Recursive Feature Elimination (RFE)
RFE is a feature selection technique used to identify the most important features in a dataset for building a predictive model.It starts by training a machine learning model on the initial set of features.
It then assigns weights to each feature based on their importance.
The least important feature(s) are then removed.
This process is repeated until the desired number of features is reached.


```python
from sklearn.feature_selection import RFE
x_sample = x.sample(n=1000, random_state=42)
y_sample = y[x_sample.index]
feature_names = x_sample.columns.tolist()
estimator = LogisticRegression(max_iter=1000, random_state=42)
n_features_to_select = 10
rfe = RFE(estimator=estimator, n_features_to_select=n_features_to_select)
rfe.fit(x_sample, y_sample)
feature_ranking = rfe.ranking_
selected_features_mask = rfe.get_support()  # Boolean mask for selected features


selected_features = [feat for feat, selected in zip(feature_names, selected_features_mask) if selected]
feature_scores = [1.0 / rank if rank == 1 else 0.0 for rank in feature_ranking]
scores_df = pd.DataFrame({
    'Feature': feature_names,
    'Score': feature_scores,
    'Ranking': feature_ranking
})
scores_df = scores_df.sort_values(by="Score", ascending=False)
scaler = MinMaxScaler()
scores_df['Normalized Score'] = scaler.fit_transform(scores_df[['Score']])

print("Selected Features:", selected_features)
print("\nScores DataFrame:")
print(scores_df)
```

#Embedded
Unlike filter methods (which assess features independently of the model) or wrapper methods (which evaluate feature subsets by repeatedly training a model), embedded methods perform feature selection as a natural part of the model training.
The model itself determines which features are most important.

Because feature selection is built into the training process, embedded methods are often more computationally efficient than wrapper methods.





```python
from sklearn.ensemble import RandomForestClassifier
x_sample = x.sample(n=1000, random_state=42)
y_sample = y[x_sample.index]
feature_names = x_sample.columns.tolist()
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_sample, y_sample)

feature_scores = rf.feature_importances_
n_features_to_select = 10  # Select top 10 features (adjust as needed)
# Create a mask for top features based on importance scores
sorted_indices = np.argsort(feature_scores)[::-1]  # Indices of features sorted by score (descending)
selected_features_mask = np.zeros(len(feature_scores), dtype=bool)
selected_features_mask[sorted_indices[:n_features_to_select]] = True

# Extract selected feature names
selected_features = [feat for feat, selected in zip(feature_names, selected_features_mask) if selected]
scores_df = pd.DataFrame({
    'Feature': feature_names,
    'Score': feature_scores
})
scores_df = scores_df.sort_values(by="Score", ascending=False)

scaler = MinMaxScaler()
scores_df['Normalized Score'] = scaler.fit_transform(scores_df[['Score']])
print("Selected Features:", selected_features)
print("\nScores DataFrame:")
print(scores_df)
```
